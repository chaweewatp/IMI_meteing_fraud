# read metering file 

```python
import pandas as pd
import numpy as np
import hashlib, json
import json
```

```python
def create_dict_measurement(df, date):
    transaction = [{key: value} for key, value in zip(list(df['Account']), list(df['Kwh_total']))]
    return {"date":date, "metering":transaction}
```

# status in 'note_read' column
0000=normal
0500= patrol meter
0600= malfunctional meter
0700= PEA.NO. missmatch
9= cannot find meter
1000= waitng for accountanct
1100= fraud metering
4400= unexpected event
7700= empty house
8800= false alarm (normal)

# retieve kWhr

```python
temp_dict={}

for Year in [2016, 2017, 2018, 2019]:
    print(Year)
    for file_name in ['meter_n3.{0:03}.csv'.format(i) for i in range(0,75)]:
        print('number of customer {}'.format(len(temp_dict)))
        print(file_name)
        df_metering = pd.read_csv('Data/Measurement/{}'.format(file_name), engine='python')
        df_temp=df_metering.drop(list(np.where(df_metering['Year']!=Year)[0])).copy()
        df_temp=df_temp.reset_index(drop=True)
        for key1, key2, key3, value in zip(list(df_temp['Account']),list(df_temp['Year']),list(df_temp['Month']),list(df_temp['Kwh_total'])):
            if key1 in temp_dict.keys():
                if key2 in temp_dict[key1]:
                    if key3 in temp_dict[key1][key2]:
                        pass
                    else:
                        temp_dict[key1][key2].update({key3:value})
                else:
                    temp_dict[key1].update({key2:{key3:value}})
            else:
                temp_dict.update({key1:{key2:{key3:value}}})
```

```python
json = json.dumps(temp_dict)
f = open("temp_dict_kWhr.json","w")
f.write(json)
f.close()
```
