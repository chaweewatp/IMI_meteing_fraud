```python
def create_dict_measurement(df, date):
    transaction = [{key: value} for key, value in zip(list(df['Account']), list(df['Kwh_total']))]
    return {"date":date, "metering":transaction}
```

```python
df_metering_all = pd.read_csv('Data/Measurement/meter_n3_000.csv', engine='python')
```

```python
df_temp=df_metering_all.drop(list(np.where(df_metering_all['Year']!=2018)[0])).copy()
df_temp=df_temp.reset_index(drop=True)
```

```python
df_temp.head()
```

```python
temp_dict={}
for key1, key2, key3, value in zip(list(df_temp['Account']),list(df_temp['Year']),list(df_temp['Month']),list(df_temp['Kwh_total'])):
    if key1 in temp_dict.keys():
        if key2 in temp_dict[key1]:
            if key3 in temp_dict[key1][key2]:
                pass
            else:
                temp_dict[key1][key2].update({key3:value})
        else:
            temp_dict[key1].update({key2:{key3:value}})
    else:
        temp_dict.update({key1:{key2:{key3:value}}})
```
